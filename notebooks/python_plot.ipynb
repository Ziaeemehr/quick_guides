```python
# Cell 6: Create a Discrete Colormap
"""
This function creates an N-bin discrete colormap from the specified 
input colormap.
"""
import matplotlib.pyplot as pl

def discrete_cmap(N, base_cmap=None):
    base = pl.colormaps.get_cmap(base_cmap) 
    color_list = base(np.linspace(0, 1, N))
    cmap_name = base.name + str(N)
    return pl.cm.colors.ListedColormap(color_list, name=cmap_name)

# Example usage
cmap = discrete_cmap(5, 'viridis')
plt.imshow(np.random.rand(10, 10), cmap=cmap)
plt.colorbar(label="Color Intensity")  # Add a label to the colorbar
plt.title("Discrete Colormap Example")  # Add a title to the plot
plt.show()
```

```python
# Binarize Data
"""
This function binarizes data based on a given threshold.
"""
def binarize(data, threshold):
    upper, lower = 1, 0
    data = np.where(data >= threshold, upper, lower)
    return data

# Example usage
data = np.random.rand(5, 5)
threshold = 0.5
binary_data = binarize(data, threshold)
print("Original Data:\n", data)
print("Binarized Data:\n", binary_data)
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Function to create a grid of subplots
def make_grid(nrows, ncols, left, right, hspace, wspace, bottom, top):
    gs = GridSpec(nrows, ncols)
    gs.update(left=left, right=right, hspace=hspace, 
              wspace=wspace, bottom=bottom, top=top)
    ax = []
    for i in range(nrows):
        for j in range(ncols):
            ax.append(plt.subplot(gs[i, j]))
    return ax

# Example usage
fig = plt.figure(figsize=(5, 4))
axes = make_grid(2, 2, left=0.1, right=0.9, 
                 hspace=0.3, wspace=0.3, bottom=0.1, top=0.9)
for ax in axes:
    ax.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# Function to plot a matrix with a colorbar
from mpl_toolkits.axes_grid1 import make_axes_locatable

def plot_matrix(A, ax, xlabel="", ylabel="", title="", 
                xticks=[], yticks=[], colorbar_label="", **kwargs):

    im = ax.imshow(A, **kwargs)
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    cbar = plt.colorbar(im, cax=cax, ax=ax)
    cbar.set_label(colorbar_label)  # Add label to the colorbar
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.set_xticks(xticks)
    ax.set_yticks(yticks)

# Example usage
import numpy as np
fig, ax = plt.subplots()
A = np.random.rand(10, 10)
plot_matrix(A, ax, xlabel="X-axis", ylabel="Y-axis", 
            title="Matrix Plot", cmap="viridis", 
            colorbar_label="Intensity")
plt.show()
```

```python
# Import necessary libraries
from statannot import add_stat_annotation
# import ptitprince as pt
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Generalized function to create a half violin plot
def half_violinplot(x, y, data, ax=None, palette="muted", 
                    violin=True, strip=True, box=True, 
                    violin_kwargs=None, strip_kwargs=None, 
                    box_kwargs=None):
    """
    Plots a half violin plot with optional stripplot and boxplot overlays.

    Parameters:
    - x: str, column name for the x-axis (categorical variable).
    - y: str, column name for the y-axis (numerical variable).
    - data: DataFrame, the data to plot.
    - ax: matplotlib Axes, the axes to plot on. If None, creates a new 
          figure and axes.
    - palette: str or list, color palette for the plot.
    - violin: bool, whether to include the half violin plot.
    - strip: bool, whether to include the strip plot.
    - box: bool, whether to include the box plot.
    - violin_kwargs: dict, additional arguments for the half violin plot.
    - strip_kwargs: dict, additional arguments for the strip plot.
    - box_kwargs: dict, additional arguments for the box plot.

    Returns:
    - ax: matplotlib Axes, the axes with the plot.
    """
    if ax is None:
        fig, ax = plt.subplots()

    violin_kwargs = violin_kwargs or {
        "bw": 0.15, "cut": 0.9, "scale": "area", "width": 0.7, 
        "inner": None, "edgecolor": "black"}
    strip_kwargs = strip_kwargs or {
        "edgecolor": "black", "linewidth": 1.0, "size": 5, 
        "jitter": 1, "zorder": 10}
    box_kwargs = box_kwargs or {
        "color": "black", "width": 0.25, "zorder": 1, "showcaps": True, 
        "boxprops": {"facecolor": "white", "zorder": 1}, 
        "showfliers": True, "whiskerprops": {"linewidth": 3, "zorder": 10}, 
        "saturation": 1}


    # Plot the half violin plot
    # if violin:
    #     pt.half_violinplot(x=x, y=y, data=data, palette=palette, 
    #     ax=ax, orient="v", **violin_kwargs)
    
    # Plot the strip plot
    if strip:
        sns.stripplot(x=x, y=y, hue=x, data=data, 
                      palette=palette, ax=ax, orient="v",
                      legend=False, **strip_kwargs)

    # Plot the box plot
    if box:
        sns.boxplot(x=x, y=y, hue=x, data=data, 
                    palette=palette, ax=ax, orient="v",
                    legend=False, **box_kwargs)

    ax.set_xlabel("")
    return ax

# Example usage
names = ["Group A", "Group B", "Group C"]
data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flatten = lambda x: [item for sublist in x for item in sublist]
x_values = flatten(
    [[name] * len(group) for name, group in zip(names, data)])
y_values = flatten(data)
df = pd.DataFrame({"Group": x_values, "Value": y_values})

# Create the plot
fig, ax = plt.subplots(figsize=(8, 6))
half_violinplot(x="Group", y="Value", data=df, ax=ax, palette="muted")
plt.title("Half Violin Plot")
plt.show()
```

```python
# Customize the appearance of a violin plot
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for the violin plot
data = [np.random.normal(size=100) for _ in range(3)]
violin = plt.violinplot(data)

# Customize the color of violin plot parts
for partname in ('cbars', 'cmins', 'cmaxes'):
    vp = violin[partname]
    vp.set_edgecolor("red")
    vp.set_linewidth(1)

# Customize the color of the violin bodies
for body in violin['bodies']:
    body.set_facecolor("lightblue")
    body.set_edgecolor("grey")
    body.set_alpha(0.7)

plt.show()
```

```python
# Function to plot a heatmap of data
def plot_data_heatmap(data, labels, title):
    from mpl_toolkits.axes_grid1 import make_axes_locatable
    fig, ax = plt.subplots(1, figsize=(10, 2.5))
    im = plt.pcolormesh(data, alpha=0.8)
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="2%", pad=0.05)
    plt.colorbar(im, cax=cax, ax=ax)
    ticks = [t + 0.5 for t in range(len(labels))]
    ax.set_yticks(ticks, labels=labels)
    ax.set_ylabel("Brain ROI")
    ax.set_xlabel("Time (fmri frame)")
    ax.set_title(title)

# Example usage
data = np.random.rand(10, 20)
labels = [f"ROI {i}" for i in range(10)]
plot_data_heatmap(data, labels, "Time Series")
plt.show()
```

```python
# Create a custom green colormap
from matplotlib.colors import LinearSegmentedColormap

colors = [(0, 0.2, 0), (0, 1, 0)]  # Dark green to bright green
n_bins = 100
green_cmap = LinearSegmentedColormap.from_list(
    'green_spectrum', colors, N=n_bins)

# Example usage
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure()
plt.scatter(x, y, c=y, cmap=green_cmap)
plt.colorbar(label='Value')
plt.show()
```
